```python
print("Columns before cleaning:", df.columns)

df.drop_duplicates(inplace=True)

columns_to_drop = ['house_rules','license']
existing_columns = [col for col in columns_to_drop if col in df.columns]
if existing_columns:
    df.drop(existing_columns, axis=1, inplace=True)

if 'price' in df.columns:
    df['price'] = df['price'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
else:
    print("Column 'price' not found in DataFrame.")

if 'service fee' in df.columns:
    df['service fee'] = df['service fee'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
else:
    print("Column 'service fee' not found in DataFrame.")

columns_to_rename = {}
if 'price' in df.columns:
    columns_to_rename['price'] = 'price_$'
if 'service fee' in df.columns:
    columns_to_rename['service fee'] = 'service_fee_$'

if columns_to_rename:
    df.rename(columns=columns_to_rename, inplace=True)
else:
    print("Neither 'price' nor 'service fee' found for renaming.")


df.dropna(inplace=True)

if 'price_$' in df.columns:
    df['price_$'] = df['price_$'].astype(float)
else:
    print("Column 'price_$' not found after renaming.")

if 'service_fee_$' in df.columns:
    df['service_fee_$_'] = df['service_fee_$'].astype(float)
else:
     print("Column 'service_fee_$' not found after renaming.")


df['id'] = df['id'].astype(str)
df['host id'] = df['host id'].astype(str)
df['last review'] = pd.to_datetime(df['last review'])
if 'Construction year' in df.columns:
    df['Construction year'] = df['Construction year'].astype(int)
else:
    print("Column 'Construction year' not found.")


if 'neighbourhood group' in df.columns:
    df.loc[df['neighbourhood group']== 'brookln','neighbourhood group']== 'Brooklyn'
else:
    print("Column 'neighbourhood group' not found.")

if 'availability 365' in df.columns:
    df=df.drop(df[df['availability 365']>500].index)
else:
     print("Column 'availability 365' not found.")
```

```python
property_types= df['room type'].value_counts().to_frame()
property_types
```

```python
import matplotlib.pyplot as plt

property_types = df['room type'].value_counts()

room_type_bar = plt.bar(property_types.index, property_types.values)

plt.bar_label(room_type_bar, labels=property_types.values, padding=4)
plt.ylim([0, 50000])
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Room Types and Their Counts in the Dataset')
plt.show()
```

```python
hood_group = df['neighbourhood group'].value_counts().to_frame()
hood_group
```

```python
avg_price = df.groupby('neighbourhood group')['price_$'].mean().sort_values(ascending=False).to_frame()
avg_price_bar = plt.bar(avg_price.index, avg_price['price_$'])
plt.bar_label(avg_price_bar, labels=round(avg_price['price_$'], 2), label_type='edge', padding=4)
plt.ylim([0, 700])
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price per Listings ($)')
plt.xticks(rotation=45)
plt.title('Average Price per Listings ($) in each Neighborhood Group')
plt.show()
```

```python
df.groupby(df['Construction year'])['price_$'].mean().to_frame().plot();
plt.xlabel('Construction Year');
plt.ylabel('Average Price ($)');
plt.title('Average Price($) for Properties in each Construction Year');
```

```python
hosts = df.groupby('host name')['calculated host listings count'].sum().sort_values(ascending = False).nlargest(10).to_frame()
hosts_bar = plt.bar(hosts.index, hosts.loc[:, 'calculated host listings count' ])
plt.bar_label(hosts_bar, label = hosts.loc[:, 'calculated host listings count' ], label_type = 'edge', padding = 3)
plt.xlabel('hosts name')
plt.ylabel('calculated Host Listings Count')
plt.xticks(rotation = 80)
plt.ylim([0, 120000])
plt.title('Top 10 Hosts by Calculated host Listings Count')
```

```python
review = df.groupby('host_identity_verified')['review rate number'].mean().sort_values(ascending=False).to_frame()
review
```

```python
review_bar = plt.bar(review.index, review['review rate number'])

plt.bar_label(review_bar, labels=round(review['review rate number'], 2), padding=4)
plt.xlabel('Host Identity Verified')
plt.ylabel('Average Review Rate Number')
plt.title('Average Review Rate by Host Identity Verification Status')
plt.ylim([0, 4])
plt.show()
```
